
# Creating a Choropleth Map of Athletes by Country

1. Importing Plotly Express:

The code imports plotly.express as px for creating interactive visualizations.

2. Replacing Country Names:

The df1['country'].replace() function is used to standardize country names by replacing "Türkiye" with "Turkey" and "Centr Afric Re" with "Central African Republic." This ensures consistency in the data for mapping.

3. Counting Athletes by Country:

The code counts the number of athletes from each country using value_counts() and resets the index to create a DataFrame called country_counts.
The columns of this DataFrame are renamed to 'Country' and 'Number of Athletes' for clarity.

4. Creating the Choropleth Map:

The choropleth map is created using px.choropleth() with the following parameters:
* locations='Country' specifies the country column for mapping.

* locationmode='country names' indicates that the country names will be used for locating the data.

* color='Number of Athletes' sets the fill color based on the number of athletes.

* hover_name='Country' allows the country name to be displayed when hovering over the map.

* title='Number of Athletes from Each Country' sets the title of the map.

* color_continuous_scale=px.colors.sequential.Plasma chooses a color scale for the choropleth.

5. Displaying the Map:

fig.show() renders the interactive choropleth map.

**Note:**

1. One must replace "Türkiye" to "Turkey" and "Centr Afric Re" with "Central African Republic." Otherwise, they won't display on the Choropleth map.

2. Russian and Belarusian athletes are forbidden from competing under their countries' flags due to the war in Ukraine. They compete as "AIN". Therefore, the Russian Federation and Belarus won't be displayed on the Choropleth map.

3. Not all countries are shown on the Choropleth map.

```python
import plotly.express as px

# Replace "Türkiye" with "Turkey" and "Centr Afric Re" with "Central African Republic"
df['country'] = df['country'].replace({
    'Türkiye': 'Turkey',
    'Centr Afric Re': 'Central African Republic'
})

# Count the number of athletes from each country
country_counts = df['country'].value_counts().reset_index()
country_counts.columns = ['Country', 'Number of Athletes']

# Create the choropleth map
fig = px.choropleth(country_counts,
                    locations='Country',
                    locationmode='country names',
                    color='Number of Athletes',
                    hover_name='Country',
                    title='Number of Athletes from Each Country',
                    color_continuous_scale=px.colors.sequential.Plasma)

# Show the map
fig.show()
```

# Cleaning and Visualizing Participant Counts per Discipline

1. Defining a Cleaning Function:

The clean_discipline function takes a discipline string, evaluates it as a Python literal using ast.literal_eval, and returns the first element. This is useful for processing disciplines that might be stored in a list format.

2. Applying the Cleaning Function:

The function is applied to the disciplines column of df1, and the cleaned results are stored in a new column called cleaned_disciplines.

3. Counting Disciplines:

discipline_counts = df1['cleaned_disciplines'].value_counts() counts the occurrences of each cleaned discipline, resulting in a series with discipline names as the index and their respective counts.

4. Creating the Bar Plot:

* plt.figure(figsize=(12, 8)) sets the figure size for the plot.

* plt.title() adds the title "Number of Participants per Discipline."

* sns.barplot() generates a horizontal bar plot with discipline_counts.values as the x-axis (number of participants) and discipline_counts.index as the y-axis (disciplines), using the 'crest' color palette.

5. Labeling the Axes:

plt.xlabel() and plt.ylabel() provide labels for the axes.

6. Displaying the Plot:

plt.show() renders the plot.

```python
def clean_discipline(discipline):
    return ast.literal_eval(discipline)[0]

# Apply the function to clean the 'disciplines' column
df['cleaned_disciplines'] = df['disciplines'].apply(clean_discipline)

# Count the occurrences of each discipline
discipline_counts = df['cleaned_disciplines'].value_counts()

# Create a DataFrame for plotting
discipline_counts_df = pd.DataFrame({
    'Discipline': discipline_counts.index,
    'Count': discipline_counts.values
})

# Sort the DataFrame by Count in descending order
ordered_counts_df = discipline_counts_df.sort_values(by='Count', ascending=False)

# Print the ordered list
print(ordered_counts_df)
```

**Merging DataFrames and Counting Athletes by Region**

1. Merging DataFrames:

* The code uses pd.merge() to combine df1 (the original athlete data) and df_regions (the country-region mapping) based on the 'country' column in df1 and the 'Country' column in df_regions. The how='left' argument ensures that all rows from df1 are retained, along with matching rows from df_regions.

2. Grouping and Counting Athletes:

* The merged DataFrame is grouped by 'Region' using merged_df.groupby('Region'). The .size() function counts the number of athletes in each region.

* The result is reset to a new DataFrame with reset_index(name='Number of Athletes'), where the counts are stored in a column named 'Number of Athletes'.

3. Displaying the Result:

print(athletes_per_region) outputs the DataFrame, showing the number of athletes from each region.


```python
merged_df = pd.merge(df, df_regions, left_on='country', right_on='Country', how='left')

# Group by 'Region' and count the number of athletes
athletes_per_region = merged_df.groupby('Region').size().reset_index(name='Number of Athletes')

# Display the result
print(athletes_per_region)
```

```python
dfv['sports'] = dfv['sports'].apply(lambda x: ast.literal_eval(x))
dfv = dfv.explode('sports').reset_index(drop=True)
dfv = dfv.rename(columns={'sports': 'sport'})

dfv.head()
```

```python
dfv['date_start'] = pd.to_datetime(dfv['date_start'])
dfv['date_end'] = pd.to_datetime(dfv['date_end'])

dfv['min_date_start'] = dfv.groupby('sport')['date_start'].transform('min')
dfv['max_date_end'] = dfv.groupby('sport')['date_end'].transform('max')

dfv['running_days'] = (dfv['date_end'] - dfv['date_start']).dt.days
dfv['sport_running_days'] = (dfv['max_date_end'] - dfv['min_date_start']).dt.days

dfv.head()
```

```python
dfv = dfv.drop(['tag', 'url'], axis=1)

dfv.head()
```

# 1. Venues hosting multiple sports

```python
venues_count = dfv['venue'].value_counts()
venues_count = venues_count[venues_count > 1]

plt.figure(figsize=(10,6))
plt.title('Venues Hosting Multiple Sports')
sns.barplot(x=venues_count, y=venues_count.index, palette='crest')
plt.xlabel('Sports Count')
plt.show()
```

# 2. Which Sports are being hosted at multiple venues?

```python
sports_count = dfv['sport'].value_counts()
sports_count = sports_count[sports_count > 1]

plt.figure(figsize=(10,6))
plt.title('Sports being Hosted in Multiple Venues')
sns.barplot(x=sports_count, y=sports_count.index, palette='flare')
plt.xlabel('Venues Count')
plt.show()
```

# 3. Which are the venues with longest running events in days

```python
df_vd  = dfv[['venue', 'running_days']].drop_duplicates().sort_values('running_days', ascending=False)
df_vd = df_vd[df_vd['running_days'] > 13]

plt.figure(figsize=(10,6))
plt.title('Venues with Longest Running Events (in Days)')
sns.barplot(x='running_days', y='venue', data=df_vd, palette='crest')
plt.xlabel('Days Count')
plt.show()
```

# 4. Which are the sports acquiring most of the days?

```python
df_sd = dfv[['sport', 'sport_running_days']].drop_duplicates().sort_values('sport_running_days', ascending=False)
df_sd = df_sd[df_sd['sport_running_days'] > 14]

plt.figure(figsize=(10,6))
plt.title('Sports being Played for Long (in Days)')
sns.barplot(x='sport_running_days', y='sport', data=df_sd, palette='flare')
plt.xlabel('Days Count')
plt.show()
```
